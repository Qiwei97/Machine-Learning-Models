```
import os
from pyabsa import ATEPCCheckpointManager, available_checkpoints
os.environ['PYTHONIOENCODING'] = 'UTF8'

available_checkpoints()
```

# Load Model

```
# 'Training Dataset': 'Rest 14, 16, Laptop 14, MAMS, twitter, tshirt, television',
# 'Training Model': 'LCF-ATEPC'

sentiment_map = {0: 'Negative', 1: 'Neutral', 2: 'Positive', -999: ''}

aspect_classifier = ATEPCCheckpointManager.get_aspect_extractor(checkpoint='english',
                                                                auto_device=True,
                                                                sentiment_map=sentiment_map)
```

# Inference

```
text = ['The apples and oranges are spoilt',
        'U.K. Considers Blocking Nvidia Takeover of Arm Over Security']

result = aspect_classifier.extract_aspect(inference_source=text,
                                          save_result=False,
                                          print_result=False,
                                          pred_sentiment=True)
```

```
import pandas as pd

result = pd.DataFrame(result)
result = result[['sentence', 'aspect', 'sentiment']]
result.sentiment = result.sentiment.map(lambda x: [sentiment_map[i] for i in x])

result
```

```
import gradio as gr
import pandas as pd

def inference(text):

  result = aspect_classifier.extract_aspect(inference_source=[text],
                                            save_result=False,
                                            print_result=False,
                                            pred_sentiment=True)
  
  result = pd.DataFrame({
      'aspect': result[0]['aspect'],
      'sentiment': result[0]['sentiment']
      })
  result.sentiment = result.sentiment.map(sentiment_map)

  return result

iface = gr.Interface(
    fn=inference, 
    inputs=["text"], 
    outputs="dataframe")

iface.launch()
```
